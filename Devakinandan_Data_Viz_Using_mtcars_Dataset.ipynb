# Data Visualization Using `plotnine`

The Palmer Penguins dataset had a variety of column types with which to explore some nice visualizations. The `mtcars` dataset is another popular dataset for doing some simple data work, but does not contain the same types of variables.

Run the following code to load the `mtcars` dataset and explore the observations and variables contained within. To learn more about this dataset check out [this site](https://stat.ethz.ch/R-manual/R-devel/library/datasets/html/mtcars.html)

```python
import statsmodels.api as sm
import pandas as pd

mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
```

The statement "the `mtcars` dataset does not contain the same types of variables as the penguins dataset" is a little true and a little false. There are no variables that contain text values, BUT there are variables that it makes sense to consider categorical variables. In most situations, it makes sense to treat the values of categorical variables as text values instead of numeric values.

Use the `astype()` method to convert the categorical variables of the `mtcars` (df) dataset to have text values in the code chunk below. We've done one for you!

```python
df["am"] = df["am"].astype(str)

# Convert the other variables below

df["cyl"] = df["cyl"].astype(str)
df["vs"] = df["vs"].astype(str)
df["gear"] = df["gear"].astype(str)
df["carb"] = df["carb"].astype(str)
```

# Converting the am variable back to float values and then trying to create this same plot

```python
df["am"] = df["am"].astype(float)
from plotnine import geom_line, ggplot, aes, geom_boxplot
from plotnine import facet_wrap


(ggplot(df, aes(x = "am", y = "mpg"))
+ geom_boxplot(fill = "Blue")
)
```

# Interpretation

## The box plot above treats am as a continuous variable for its values ranging from 0 to 1. This does not make sense in a realistic context as am should be treated as a categorical variable

```python
# Create plot for (2) here
from plotnine import geom_density, geom_histogram
(ggplot(df,
aes(
  x = "hp",
  fill = "cyl"
))
+ geom_histogram()
)
```

```python
df["cyl"] = df["cyl"].astype(float)
from plotnine import geom_density, geom_histogram
(ggplot(df,
aes(
  x = "hp",
  fill = "cyl"
))
+ geom_histogram()
)
```

# Difference

## In the first plot, the variable cyl is treated as categorical. It has three distinct categories, represented by different colors: 4, 6, and 8 cylinders

## In the second plot, the variable cyl has been converted to a continuous float. As a result, it is no longer treated as a categorical variable

### Since the fill aesthetic is based on a continuous variable now, it results in a single-color histogram without any differentiation between different cylinder counts. The plot only shows the distribution of hp values without distinguishing between different values of cyl

Some of these variables, like the `cyl` variable, have numeric values that actually make sense as numbers (i.e. the number of cylinders in the engine). However, it doesn't make the most sense to "do math" with this type of variable (e.g. take averages and such) because there are so few different value this can take on AND they're an explicit choice made by the car manufacturer. So, it makes more sense to treat the `cyl` variable as a categorical variable despite it having numeric values.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create plot with first new geometry here
df = sns.load_dataset('mpg').dropna()
plt.scatter(df['horsepower'], df['mpg'], color='green', alpha=0.6)
plt.title('Relationship between Horsepower and MPG')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Gallon (MPG)')
plt.grid(True)
```

### As horsepower increases, the MPG generally decreases, which implies a negative correlation between the horsepower and mpg. Higher horsepower is associated with lower fuel efficiency

```python
# Create plot with second new geometry here
plt.figure(figsize=(8,6))
plt.hist(df['mpg'], bins=10, color='blue', alpha=0.7)
plt.title('Distribution of Miles per Gallon (MPG)')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.grid(True)
```

### The above histogram shows the distribution of miles per gallon (MPG). Most cars are usually within the 15-30 MPG range, with a denser concentration between 15 and 25 MPG, thus implying that a majority of vehicles fall in this range

```python
# Create plot with third new geometry here
from mpl_toolkits.mplot3d import Axes3D

# Creating a 3D scatter plot using horsepower, mpg, and weight
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['horsepower'], df['mpg'], df['weight'], c=df['mpg'], cmap='viridis', marker='o')

ax.set_xlabel('Horsepower')
ax.set_ylabel('Miles per Gallon (MPG)')
ax.set_zlabel('Weight')

plt.title('3D Scatter Plot of Horsepower, MPG, and Weight')

```

### The 3D scatter plot shows the relationship between horsepower, miles per gallon (MPG), and weight. The data points can be color diffeentiated based on their MPG values, with the vertical axis representing the vehicle weight. As horsepower increases, lower the MPG and higher is the vehicle weight
