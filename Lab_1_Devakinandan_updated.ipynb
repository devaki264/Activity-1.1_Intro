```python
# Defining the specific colors for the regions based on 'four_regions'
color_mapping = {
    'asia': 'red',
    'africa': 'aqua',
    'americas': 'lime',
    'europe': 'yellow'
}
data = pd.read_csv('q1data.csv')
# Assuming you already have a dataset loaded
filtered_data_q1 = data.dropna(subset=['income', 'life_exp', 'population', 'four_regions'])

# Creating the plot
plot = (
    ggplot(filtered_data_q1, aes(x='income', y='life_exp', size='population', fill='four_regions')) +

    # Adding "2010" watermark in the background
    annotate('text', x=4000, y=50, label='2010', size=190, alpha=0.1, color='grey') +

    # Adding points for the countries
    geom_point(alpha=0.7, color="black", stroke=0.5) +

    # Scaling the bubble size based on population
    scale_size_continuous(range=(3, 22)) +

    # Adjusting X-axis (Income) to log scale
    scale_x_log10(limits=(500, 65000), breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000]) +

    # Adjusting the life expectancy
    scale_y_continuous(limits=(20, 90), breaks=range(20, 100, 10)) +

    # Applying color mapping for the regions
    scale_fill_manual(values=color_mapping) +

    # Adding Income Level annotations and diamond markers
    annotate('text', x=1300, y=90, label='INCOME LEVEL 1', size=8, color='black') +
    annotate('text', x=5000, y=90, label='LEVEL 2', size=8, color='black') +
    annotate('text', x=14000, y=90, label='LEVEL 3', size=8, color='black') +
    annotate('text', x=48000, y=90, label='LEVEL 4', size=8, color='black') +


    annotate('point', x=3000, y=90, size=5, color='black', shape='D') +
    annotate('point', x=8000, y=90, size=5, color='black', shape='D') +
    annotate('point', x=24000, y=90, size=5, color='black', shape='D') +

    # Using minimal theme
    theme_minimal() +
    labs(x='Income', y='Life expectancy') +
    guides(size=False, fill=False) +
    # Adjust figure size
    theme(figure_size=(11, 6))
)

# Show the plot
plot.show()
```

```python
data=pd.read_csv('/content/q2data.csv')
```

```python
from plotnine import ggplot, aes, geom_point, scale_size_continuous, theme_minimal, labs, scale_x_continuous, scale_y_continuous, guides, scale_fill_manual, annotate


# Define the specific colors for the bubbles
color_mapping = {
    'asia': 'red',      # Red for Asia
    'africa': 'aqua',   # Aqua for Africa
    'americas': 'lime', # Lime for Americas
    'europe': 'yellow'  # Yellow for Europe
}

# Filter the dataset for only the four specified regions
filtered_data = data[data['four_regions'].isin(color_mapping.keys())]

# Ensure that the 'energy' column is numeric and handle missing values
filtered_data['energy'] = pd.to_numeric(filtered_data['energy'], errors='coerce').fillna(0)

# Create the plot using plotnine
plot = (
    ggplot(filtered_data, aes(x='exports', y='imports', size='energy', fill='four_regions')) +
    annotate('text', x=110, y=200, label='1997', size=150, alpha=0.1, color='grey') +  # Center and enlarge "1997"
    geom_point(alpha=0.7, color="black", stroke=0.5) +  # Add light black outline and keep color fills
    scale_size_continuous(range=(2, 12)) +
    scale_x_continuous(breaks=range(0, 240, 20), labels=lambda l: ["{}".format(v) for v in l], limits=(0, 220)) +  # Remove percentage symbol
    scale_y_continuous(breaks=range(0, 450, 50), limits=(0, 400)) +  # Set Y-axis breaks and limits
    scale_fill_manual(values=color_mapping) +  # Apply the custom fill color mapping
    theme_minimal() +
    labs(x='Exports ', y='Imports') +
    guides(size=False, fill=False)  # Disable legends for size and color
)

# Show the plot
plot.show()
```

```python
from plotnine import ggplot, aes, geom_point, scale_size_continuous, theme_minimal, labs, scale_x_continuous, scale_y_log10, guides, scale_fill_manual, annotate, theme
import pandas as pd

# Define the specific colors for the regions based on 'four_regions'
color_mapping = {
    'asia': 'red',
    'africa': 'aqua',
    'americas': 'lime',
    'europe': 'yellow'
}

# Assuming you already have a dataset loaded
data = pd.read_csv('q3data.csv')

# Filter the dataset to remove missing values in relevant columns
filtered_data_q3 = data.dropna(subset=['internet_users', 'gdp', 'income', 'four_regions'])

# Create a plot using plotnine
plot = (
    ggplot(filtered_data_q3, aes(x='internet_users', y='gdp', size='income', fill='four_regions')) +

    # Adjust the "2001" watermark: make it bigger and centered
    annotate('text', x=45, y=2800, label='2001', size=190, alpha=0.1, color='grey') +  # Big "2001" watermark

    geom_point(alpha=0.7, color="black", stroke=0.5) +  # Add bubbles with a black outline
    scale_size_continuous(range=(3, 18)) +  # Increase the size range for larger bubbles
    scale_x_continuous(limits=(0, 100), breaks=range(0, 101, 10)) +  # Adjust X-axis to 0-100 for Internet users
    scale_y_log10(limits=(200, 100000), breaks=[200, 500, 1000, 5000, 10000, 20000, 50000, 100000]) +  # Set Y-axis limits
    scale_fill_manual(values=color_mapping) +  # Apply color mapping for regions
    theme_minimal() +  # Use minimal theme
    labs(x='Individuals using the Internet', y='GDP/capita') +  # Label the axes
    guides(size=False, fill=False) +  # Disable legends for size and color
    theme(figure_size=(11, 5))  # Set figure size using theme
)

# Show the plot
plot.show()
```

```python
from plotnine import ggplot, aes, geom_bar, scale_fill_manual, theme_minimal, labs, theme
import pandas as pd

# Load your dataset
data = pd.read_csv('q3data.csv')

# Filter the dataset to remove missing values in relevant columns
filtered_data_q3 = data.dropna(subset=['internet_users', 'gdp', 'income', 'four_regions'])

# Create a plot using plotnine with geom_bar
plot = (
    ggplot(filtered_data_q3, aes(x='four_regions', fill='four_regions')) +

    # Bar plot to show count of observations by region
    geom_bar() +
    scale_fill_manual(values=color_mapping) +  # Apply color mapping for regions
    theme_minimal() +  # Use minimal theme
    labs(x='World Regions', y='Count of Observations') +  # Label the axes
    theme(figure_size=(11, 5))  # Set figure size using theme
)

# Show the plot
plot.show()
```

### Geom_bar(): Useful for aggregating data into categorical bins.



*   Why it's good: If you want to summarize and compare grouped data, for instance, showing the average GDP for different world regions.


*   Why it's  bad: It loses individual point-level detail, so its not ideal for scatter-type relationships
